# Cómo se procesa una imagen en un ordenador

Antes de ponernos a hablar de como usar redes convolucionales para el procesado de imágenes vamos a dedicarle un tiempo a ver cómo se procesa una imagen en un ordenador

## Imágenes en blanco y negro

Vamos a descargarnos el dataset de `MNIST`, que es un dataset de imágenes en blanco y negro de dígitos escritos a mano. Vamos a empezar por aquí parea ver cómo se procesa una imagen en blanco y negro y luego pasamos a imágenes a color

```python
from torchvision import datasets

data = datasets.MNIST(
    root="data",
    train=True,
    download=True
    # transform=ToTensor()
)

len(data)
```

```python
sample_data = data[0]
type(sample_data), len(sample_data)
```

Vemos que la muestra es una tupla de 2 items, esto es porque tenemos la imagen y la etiqueta. Ahora la etiqueta no nos interesa, así que nos quedamos con la imagen

```python
sample_img = sample_data[0]
display(sample_img)
type(sample_img)
```

Como podemos ver tenemos una imagen de un 5 y es de tipo `PIL`, que es una librería de imágenes de Python. Vamos a pasarla a un array de Numpy para ver cómo se representa

```python
import numpy as np

sample_array = np.array(sample_img)
sample_array.shape
```

Vemos que internamente se representa como una matriz de 28x28 píxeles

Vamos ahora a dibujarlo con una utilidad muy gráfica de la librería Pandas

```python
import pandas as pd

sample_df = pd.DataFrame(sample_array)
sample_df.style.set_properties(**{'font-size':'6pt'}).background_gradient('Greys')
```

Podemos ver que cada pixel es representado como una intensidad de blanco y negro

# Imágenes a color

Ya hemos visto cómo se representa una imagen en blanco y negro, para representar una imagen en color, lo que se hace es replicar la matriz que hemos visto antes 3 veces, así la primera de las matrices corresponderá al matiz rojo de la imagen, la segunda al matiz verde y la tercera al azúl (RGB)

Vamos a crear ahora una matriz de 3 canales asignando la primera dimensión a la matriz anterior y las otras dos dimensiones todo ceros

```python
zeros = np.zeros_like(sample_array)
sample_array_red = np.dstack((sample_array, zeros, zeros))
zeros.shape, sample_array_red.shape
```

```python
from PIL import Image

sample_img_red = Image.fromarray(sample_array_red.astype(np.uint8))
display(sample_img_red)
```

Ahora una matriz de 3 canales asignando la segunda dimensión a la matriz anterior y las otras dos dimensiones todo ceros

```python
sample_array_green = np.dstack((zeros, sample_array, zeros))
sample_img_green = Image.fromarray(sample_array_green.astype(np.uint8))
display(sample_img_green)
```

Ahora una matriz de 3 canales asignando la tercera dimensión a la matriz anterior y las otras dos dimensiones todo ceros

```python
sample_array_blue = np.dstack((zeros, zeros, sample_array))
sample_img_blue = Image.fromarray(sample_array_blue.astype(np.uint8))
display(sample_img_blue)
```

Por último una matriz de 3 canales asignando todas las dimensiones a la matriz anterior

```python
sample_array_3d = np.dstack((sample_array, sample_array, sample_array))
sample_img_3d = Image.fromarray(sample_array_3d.astype(np.uint8))
display(sample_img_3d)
```
