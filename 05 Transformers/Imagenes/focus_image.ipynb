# Focuss on Input Embedding

```python
color_multi_head_attention = (254,225,187)
color_embedding = (252,225,225)
color_add_norm = (241,243,193)
color_lineal = (219,223,239)
color_softmax = (205,231,207)
color = color_multi_head_attention
margin = 10
```

```python
margin = 10
img_rgba = img.convert("RGBA")
data = img_rgba.getdata()
new_data = []
transparence = 150
for i, pixel in enumerate(data):
    if pixel[0] < color[0] + margin and pixel[0] > color[0] - margin and \
        pixel[1] < color[1] + margin and pixel[1] > color[1] - margin and \
        pixel[2] < color[2] + margin and pixel[2] > color[2] - margin:
        new_data.append(pixel)
    else:
        new_pixel = (pixel[0], pixel[1], pixel[2], transparence)
        new_data.append(new_pixel)
img_rgba.putdata(new_data)
img_rgba
```
