$$lr = d_{model}^{-0.5} \cdot min \left(stepNumber^{-0.5}, stepNumber \cdot stepWarmUp^{-1.5} \right)$$

```python
dim_embeding_model = 512
```

```python
import numpy as np
import matplotlib.pyplot as plt

warmup_steps = 4000
step_num = np.arange(0, 1000000)
lr = np.power(dim_embeding_model, -0.5) * np.minimum(np.power(step_num, -0.5), step_num * np.power(warmup_steps, -1.5))

print(lr.max(), lr.min())

plt.plot(step_num, lr)
plt.xlabel('step_num')
plt.ylabel('lr')
plt.show()
```
